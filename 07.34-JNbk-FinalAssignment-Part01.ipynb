Import two necessary libraries: Pandas and Numpy.

```python
import numpy as np 
import pandas as pd
```

Read the dataset into a Pandas dataframe.

```python
df_survey = pd.read_csv('https://cocl.us/datascience_survey_data')
df_survey.rename(columns={'Unnamed: 0':'Topic'}, inplace=True)
df_survey.set_index('Topic', inplace=True)
display(df_survey)
```

```python
df_survey.sort_values(['Very interested'], ascending=False, axis=0, inplace=True)
df_survpor= round(100 * df_survey / 2233, 2)
df_survpor
```

Import necessary libraries for creating charts, graphs, etc.

```python
import matplotlib
import matplotlib.pyplot as plt
```

Lists of data to be used in grouped bar chart.

```python
labels = list(df_survpor.index)
vInt = list(df_survpor['Very interested'])
sInt = list(df_survpor['Somewhat interested'])
nInt = list(df_survpor['Not interested'])
```

```python
# Create plot object
ax0 = df_survpor.plot(kind='bar',
                    figsize=(20, 8),
                    width=0.8,                  
                    color=('#5cb85c', '#5bc0de', '#d9534f'),
                    fontsize=14
                   )

# Control boder visibility
ax0.spines['top'].set_visible(False)
ax0.spines['right'].set_visible(False)
ax0.spines['left'].set_visible(False)

# Control axes, labels and title
ax0.axes.get_yaxis().set_visible(False)
#ax0.set_xticklabels(new_labels, fontsize=your_fontsize)
ax0.axes.get_xaxis().get_label().set_visible(False)
ax0.set_title('Percentage of Respondents\' Interest in Data Science Areas', fontsize=16)
ax0.legend(list(df_survpor.columns), loc='upper right', fontsize='x-large')

# Control bars annotation
for p in ax0.patches:
    ax0.annotate(str(p.get_height())+'%', 
                 (p.get_x(), p.get_height()), 
                 ha='left',
                 va='center',
                 xytext=(0, 10),
                 textcoords='offset points',
                 fontsize=14)
```
